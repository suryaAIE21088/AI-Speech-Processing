```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
audio_path = r"C:/Users\pujit\Downloads\surya.wav"
audio_signal, sample_rate = librosa.load(audio_path)
plt.figure(figsize=(12, 6))
librosa.display.waveshow(audio_signal, sr=sample_rate)
plt.title('Customized Speech Signal Waveform')
plt.show()
```

```python
print("Length of the signal:", len(audio_signal), "samples")
print("Sampling rate:", sample_rate, "Hz")
print("Magnitude range:", np.min(audio_signal), "to", np.max(audio_signal))
```

```python
import IPython.display as ipd
start_sample = int(2 * sample_rate)
end_sample = int(6 * sample_rate)
audio_segment = audio_signal[start_sample:end_sample]
librosa.display.waveshow(audio_segment, sr=sample_rate)
ipd.Audio(audio_segment, rate=sample_rate)
```

```python
segment_1_start = int(1.5 * sample_rate)
segment_1_end = int(3 * sample_rate) 
segment_2_start = int(3.5 * sample_rate)
segment_2_end = int(5 * sample_rate)   
segment_1 = audio_signal[segment_1_start:segment_1_end]
segment_2 = audio_signal[segment_2_start:segment_2_end]
print("Playing Segment 1:")
plt.figure(figsize=(12, 6))
librosa.display.waveshow(segment_1, sr=sample_rate)  
plt.title('Segment 1')
plt.show()
ipd.Audio(segment_1, rate=sample_rate)
```

```python
print("Playing Segment 2:")
plt.figure(figsize=(12, 6))
librosa.display.waveshow(segment_2, sr=sample_rate)  
plt.title('Segment-2')  
plt.show()
ipd.Audio(segment_2, rate=sample_rate)
```
